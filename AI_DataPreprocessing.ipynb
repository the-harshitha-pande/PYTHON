```python
#loading csv file
import pandas as pd

df = pd.read_csv("default of credit card clients.csv", header=1)
df.head()
```

```python
#🔍Check dataset
print(df.shape)
print(df.columns)
```

```python
df["default payment next month"].head()
```

```python
#Clean column names (IMPORTANT Sometimes column name has spaces → fix it:)
df.columns = df.columns.str.strip()
```

```python
#Reduce dataset to 4000 from 30,000
# Separate classes
fraud = df[df['default payment next month'] == 1]
normal = df[df['default payment next month'] == 0]

print("Fraud:", len(fraud))
print("Normal:", len(normal))
```

```python
#REnaming
df.rename(columns={"default payment next month": "target"}, inplace=True)
```

```python
df['target']
```

```python
# majority vs minority risk check 0->safe, 1->risk
df['target'].value_counts()
```

```python
#RECUDE DATASET
fraud = df[df['target'] == 1]
normal = df[df['target'] == 0]

print("Fraud:", len(fraud))
print("Normal:", len(normal))
```

```python
#Take samples
fraud_sample = fraud.sample(n=1000, random_state=42)
normal_sample = normal.sample(n=3000, random_state=42)

#Combine
df_small = pd.concat([fraud_sample, normal_sample])

#Shuffle
df_small = df_small.sample(frac=1, random_state=42)

#result
print(df_small.shape)
```

```python
#Check balance
df_small['target'].value_counts()
```

```python
#Save reduced dataset
df_small.to_csv("reduced_dataset.csv", index=False)
```

```python
#Check Missing Values : If all 0 → good
df_small.isnull().sum()
```

```python
#Drop unnecessary column (ID)
df_small.drop(columns=['ID'], inplace=True)
```
